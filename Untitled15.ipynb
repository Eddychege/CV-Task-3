```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from scipy import stats
```

```python
df = pd.read_excel('/content/202250508 - Sales Operation Data Set.xlsx')
```

```python
print("First five rows of the data:")
print(df.head())
print("\nSummary of data information:")
print(df.info())
print("\nDescriptive satatistics of numerical columns:")
print(df.describe())
print("\nDf columns:")
print(df.columns)
```

```python
top_10_agents_gwp = df.groupby('Agent')['GWP'].sum().nlargest(10).reset_index()
top_10_agent_names = top_10_agents_gwp['Agent'].to_list()

print("Top 10 agents with the highest GWP:")
print(top_10_agents_gwp)

top_10_df = df[df['Agent'].isin(top_10_agent_names)]

upesi = 'UPESI WORLD INSURANCE AGENCY'
other_9_agents = [agent for agent in top_10_agent_names if agent !=upesi]

print(f"\nUpesi: {upesi}")
print(f"Other 9 agents: {other_9_agents}")

upesi_df = top_10_df[top_10_df['Agent'] == upesi]
other_9_df = top_10_df[top_10_df['Agent'].isin(other_9_agents)]

print("\nData types in the Top 10 data:")
top_10_df.info()

```

EDA

 GWP COMPARISON



```python
gwp_comparison_top_10 = top_10_df.groupby('Agent')['GWP'].sum()
upesi_gwp_top_10 = gwp_comparison_top_10[upesi]
other_9_gwp_top_10_sum = gwp_comparison_top_10[other_9_agents].sum()
other_9_avg_gwp_top_10 = other_9_gwp_top_10_sum / len(other_9_agents)

print(f"\nUpesi's Total GWP (Top 10): kes {upesi_gwp_top_10:,.2f}")
print(f"Total GWP of Other 9 Agents (Top 10): kes {other_9_gwp_top_10_sum:,.2f}")
print(f"Average Total GWP o fother 9 Agents (Top 10): kes {other_9_avg_gwp_top_10}")

plt.figure(figsize=(10, 6))
gwp_comparison_top_10.sort_values(ascending=False).plot(kind='bar')
plt.title('Total GWP Comparison for Top 10 Agents')
plt.xlabel('Agent')
plt.ylabel('Total GWP')
plt.xticks(rotation=45, ha="right")
plt.show()


```

NUMBER OF POLICIES SOLD


```python
policies_sold_top_10 = top_10_df.groupby('Agent')['Policy No'].nunique()
upesi_policies_top_10 = policies_sold_top_10[upesi]
other_9_policies_top_10_sum = policies_sold_top_10[other_9_agents].sum()
other_9_avg_gwp_top_10 = other_9_policies_top_10_sum / len(other_9_agents)

print(f"\nUpesi's Total Policies Sold (Top 10): {upesi_policies_top_10}")
print(f"Total Policies Sold of Other 9 Agents (Top 10): {other_9_policies_top_10_sum}")
print(f"Average Total Policies Sold o fother 9 Agents (Top 10): {other_9_avg_gwp_top_10}")

plt.figure(figsize=(10, 6))
policies_sold_top_10.sort_values(ascending=False).plot(kind='bar')
plt.title('Total Policies Sold Comparison for Top 10 Agents')
plt.xlabel('Agent')
plt.ylabel('Total Policies Sold')
plt.xticks(rotation=45, ha="right")
plt.show()
```

AVERAGE POLICY SIZE(PREMIUM)


```python
avg_policy_size_top_10 = gwp_comparison_top_10 / policies_sold_top_10
upesi_avg_size_top_10 = avg_policy_size_top_10[upesi]
other_9_avg_size_top_10 = avg_policy_size_top_10[other_9_agents].mean()

print(f"\nUpesi's Average Policy Size (Top 10): kes {upesi_avg_size_top_10}")
print(f"Average Policy Size of Other 9 Agents (Top 10): kes {other_9_avg_size_top_10}")

plt.figure(figsize=(10, 6))
avg_policy_size_top_10.sort_values(ascending=False).plot(kind='bar')
plt.title('Average Policy Size Comparison for Top 10 Agents')
plt.xlabel('Agent')
plt.ylabel('Average Policy Size')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python


product_mix_top_10_type = top_10_df.groupby('Agent')['Type'].value_counts(normalize=True).mul(100).rename('Percentage').reset_index()

upesi_product_mix_top_10_type = product_mix_top_10_type[product_mix_top_10_type['Agent'] == upesi]
other_9_product_mix_top_10_type = product_mix_top_10_type[product_mix_top_10_type['Agent'].isin(other_9_agents)].groupby('Type')['Percentage'].mean().sort_values(ascending=False)

print("\nUpesi's Product Type Mix (Top 10) (%):\n", upesi_product_mix_top_10_type)
print("\nAverage Product Type Mix of Other 9 Agents (Top 10) (%):\n", other_9_product_mix_top_10_type)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.barplot(data=upesi_product_mix_top_10_type, x='Type', y='Percentage', ax=axes[0])
axes[0].set_title(f'{upesi}\'s Product Type Mix')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right') # Using set_xticklabels

sns.barplot(x=other_9_product_mix_top_10_type.index, y=other_9_product_mix_top_10_type.values, ax=axes[1])
axes[1].set_title('Average Product Type Mix of Other 9')
axes[1].set_xticklabels(other_9_product_mix_top_10_type.index, rotation=45, ha='right') # Using set_xticklabels

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming your main DataFrame containing the top 10 agents' data is named 'top_10_df'

# Filter data for the 9 agents (excluding Upesi)
other_9_agents_df = top_10_df[top_10_df['Agent'] != 'UPESI WORLD INSURANCE AGENCY'].copy()

# 1. Calculate Average GWP per Policy (Overall)
gwp_by_agent = other_9_agents_df.groupby('Agent')['GWP'].sum()
policy_count_by_agent = other_9_agents_df.groupby('Agent')['Policy No'].nunique()
avg_gwp_overall = gwp_by_agent / policy_count_by_agent

# 2. Calculate Average GWP per 'COMP' Policy
comp_policies = other_9_agents_df[other_9_agents_df['Cover'] == 'COMP'].groupby('Agent')['GWP'].mean().fillna(0)
avg_gwp_comp = comp_policies

# 3. Calculate New Business GWP Ratio
new_business_gwp_by_agent = other_9_agents_df[other_9_agents_df['Type'] == 'NB'].groupby('Agent')['GWP'].sum().fillna(0)
total_gwp_by_agent = other_9_agents_df.groupby('Agent')['GWP'].sum().fillna(0)
new_business_ratio = (new_business_gwp_by_agent / total_gwp_by_agent).fillna(0)

# 4. Calculate Total Number of Policies Sold
total_policies_by_agent = other_9_agents_df.groupby('Agent')['Policy No'].nunique()

# 5. Calculate Average Monthly GWP
other_9_agents_df['Production (Date)'] = pd.to_datetime(other_9_agents_df['Production (Date)'])
monthly_gwp_by_agent = other_9_agents_df.groupby(['Agent', pd.Grouper(key='Production (Date)', freq='M')])['GWP'].sum().groupby('Agent').mean().fillna(0)
avg_monthly_gwp = monthly_gwp_by_agent

# Create a DataFrame for clustering features
clustering_data = pd.DataFrame({
    'Agent': avg_gwp_overall.index,
    'Avg_GWP_Overall': avg_gwp_overall.values,
    'Avg_GWP_COMP': avg_gwp_comp.values,
    'New_Business_Ratio': new_business_ratio.values,
    'Total_Policies': total_policies_by_agent.values,
    'Avg_Monthly_GWP': avg_monthly_gwp.values
})
clustering_data = clustering_data.set_index('Agent')

# Handle potential infinite values in 'New_Business_Ratio'
clustering_data.replace([float('inf'), -float('inf')], 0, inplace=True)

# Standardize the features
features_to_scale = ['Avg_GWP_Overall', 'Avg_GWP_COMP', 'New_Business_Ratio', 'Total_Policies', 'Avg_Monthly_GWP']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_data[features_to_scale])

scaled_clustering_df = pd.DataFrame(scaled_features, columns=features_to_scale, index=clustering_data.index)

print("Scaled Features for Clustering (First few rows):\n", scaled_clustering_df.head())

# Now 'scaled_clustering_df' is ready for applying clustering algorithms.
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Apply K-Means with different numbers of clusters
n_clusters_options = [3, 4, 5]
cluster_results = {}

for n_clusters in n_clusters_options:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(scaled_clustering_df)
    cluster_results[n_clusters] = clusters

# Add cluster labels back to the original (non-scaled) clustering data
clustered_data = clustering_data.copy()
for n_clusters, labels in cluster_results.items():
    clustered_data[f'Cluster_{n_clusters}'] = labels

print("Clustered Data (showing cluster assignments for different n_clusters):\n", clustered_data)

# Analyze the characteristics of each cluster (for n_clusters=3 as an example)
n_clusters_to_analyze = 3
print(f"\nMean Feature Values for Each Cluster (K={n_clusters_to_analyze}):")
print(clustered_data.groupby(f'Cluster_{n_clusters_to_analyze}')[['Avg_GWP_Overall', 'Avg_GWP_COMP', 'New_Business_Ratio', 'Total_Policies', 'Avg_Monthly_GWP']].mean())

# Optional: Visualize the clusters (for n_clusters=3, using two principal components for visualization)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_clustering_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'], index=scaled_clustering_df.index)
pca_df['Cluster'] = clustered_data[f'Cluster_{n_clusters_to_analyze}']

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=100)
plt.title(f'Agent Clusters (K={n_clusters_to_analyze}) Visualized in 2D PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# To determine the optimal number of clusters, you would typically use:
# 1. Elbow Method: Plotting the within-cluster sum of squares (WCSS) against the number of clusters.
# 2. Silhouette Analysis: Calculating the silhouette score for different numbers of clusters.

# We can implement these methods if you'd like to refine the number of clusters.
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate Within-Cluster Sum of Squares (WCSS) for different numbers of clusters
wcss = []
range_n_clusters = range(1, 9)  # Try clustering from 1 to 8 clusters
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(scaled_clustering_df)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, wcss, marker='o', linestyle='--')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()

# We will proceed with the Elbow Method result for now.
```
